```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt

# List of stock tickers for the portfolio
tickers = ['HDFCBANK.NS', 'RELIANCE.NS', 'FORTIS.NS', 'AMBUJACEM.NS', 'ITC.NS']
weights1 = 0.11770820586072048,0.021831950590664565, 0.0373192717812358, 0.812708275028356,0.01043229673902323] # Equal weights for each stock in the portfolio

# Fetch historical stock prices1 for each stock in the portfolio from Yahoo Finance
historical_prices1 = {}
for ticker in tickers:
    stock1 = yf.Ticker(ticker)
    historical_data1 = stock1.history(period="max")
    historical_prices1[ticker] = historical_data1['Close'].values

# Calculate daily returns for each stock in the portfolio
returns1 = {}
for ticker, prices in historical_prices1.items():
    returns1[ticker] = np.diff(prices) / prices[:-1]

# Define parameters for simulation
num_simulations1 = 1000
num_days1 = 100
mean_returns1 = {ticker: np.mean(ret) for ticker, ret in returns1.items()}
std_returns1 = {ticker: np.std(ret) for ticker, ret in returns1.items()}
simulations = {ticker: [] for ticker in tickers}
for ticker in tickers:
    for _ in range(num_simulations1):
        prices = [historical_prices1[ticker][-1]]
        for _ in range(num_days1):
            # Generate a random sample from a normal distribution with mean and standard deviation of daily returns
            daily_return = np.random.normal(mean_returns1[ticker], std_returns1[ticker])
            price = prices[-1] * (1 + daily_return)
            prices.append(price)
        simulations[ticker].append(prices)

# Plot simulations for each stock in the portfolio
for ticker in tickers:
    plt.figure()
    for sim in simulations[ticker]:
        plt.plot(sim)
    plt.xlabel('Days')
    plt.ylabel('Stock Price')
    plt.title(f'Monte Carlo Simulation: {ticker} Stock Price Prediction')
    plt.show()

# Calculate portfolio returns for each simulation
simulated_portfolio_values = np.zeros(num_days1)
for _ in range(num_simulations1):
    portfolio_value1 = 1
    for i in range(num_days1):
        daily_returns1 = [np.random.normal(np.mean(returns1[ticker]), np.std(returns1[ticker])) for ticker in tickers]
        daily_return_portfolio = np.dot(weights1, daily_returns1)
        portfolio_value1 *= (1 + daily_return_portfolio)
        simulated_portfolio_values[i] += portfolio_value1

# Calculate the average portfolio values
simulated_portfolio_values 

# Plot the Monte Carlo simulation for the portfolio
plt.plot(range(1, num_days1 + 1), simulated_portfolio_values)
plt.xlabel('Days')
plt.ylabel('Portfolio Value')
plt.title('Monte Carlo Simulation: Portfolio Value Prediction')
plt.show()
```
